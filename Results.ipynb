```python
dataset = [df_BB, df_BL, df_RB, df_RL, df_GS, df_GM]
models = ['Bert Base', 'Bert Large', 'Roberta Base', 'Roberta Large', 'GPT2 Small', 'GPT2 Medium']
res = []
for idx, df in enumerate(dataset):
    df['Model'] = models[idx]
    df['Ordering'] = ''
    df.at[0, 'Ordering'] = 'Original'
    df.at[1, 'Ordering'] = 'Full Shuffle'
    res.append(df)
rdf = pd.concat(res, axis=0)
rdf
```

```python
fig, ax = plt.subplots(2, 3, figsize=(20, 8))
fig.tight_layout(pad=8.0)
r, c = 0, 0
for ds in rdf.columns.difference(['Model', 'Ordering']):
    if c == 3:
        r = 1
        c = 0
    sns.barplot(data=rdf[[ds, 'Model', 'Ordering']], x = 'Model', y = ds, hue='Ordering', ax=ax[r][c])
    ax[r][c].tick_params(axis='x', rotation=45)
    if r == 0 and c == 0:
        handles, labels = ax[r][c].get_legend_handles_labels()
        fig.legend(handles, labels, loc='lower center')
    ax[r][c].legend([],[], frameon=False)
    ax[r][c].set_ylabel('Accuracy')
    ax[r][c].set_xlabel('')
    ax[r][c].set_ylim([0, 1])
    ax[r][c].set_title(ds)
    c += 1

fig.savefig('barplot.png')
```
